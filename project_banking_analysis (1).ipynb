### Examine the distribution of unique categories in categorial columns 

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']
data['Income Band'] = pd.cut(data['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
data['Income Band'].value_counts().plot(kind = 'bar')
```

```python
categorical_cols = ['BRId', 'GenderId', 'Amount of Credit Cards', 'Nationality','Occupation', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting']
for col in categorical_cols:
    print(f'Value counts for {col}:')
    display(data[col].value_counts())
```

## Univariate analysis
Univariate analysis is a type of data analysis that examines each variable in a dataset independently. Its primary purpose is to describe and summarize the main features of a single variable, typically through statistical measures and visualizations.


```python
for i, predictor in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=data, x=predictor )
    plt.xlabel(predictor)
    plt.tight_layout()
    plt.show()
```

## Bivariate analysis
Bivariate analysis involves analyzing two variables simultaneously to understand the relationship between them. It helps identify patterns, associations, correlations, or differences between the two variables.

```python
for i, predictor in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=data, x=predictor, hue='Nationality' )
    plt.xlabel(predictor)
    plt.tight_layout()
    plt.show()
```

### Histplot of value counts for different occupation 


```python
for col in categorical_cols:
    if col == 'occupation':
        continue
    plt.figure(figsize=(8,6))
    sns.histplot(data[col])
    plt.title('Histogram of occupation count')
    plt.xlabel(col)
    plt.ylabel('count')
    plt.show()
        
```

## Numerical analysis

```python
numerical_cols = ['Estimated Income','Superannuation Savings', 'Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

#univariate analysis and visulization
for i, col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(data[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
    
```

## Heatmaps 

```python
numerical_cols = ['Estimated Income','Superannuation Savings', 'Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

correlation_matrix = data[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
